Section A – Concept Application


Q1. Fashion MNIST contains raw 28×28 pixel images. Why is Deep Learning more suitable than traditional ML for classifying raw image data?

Answer:

Fashion MNIST contains 28×28 grayscale images of clothing items. Deep Learning is more suitable because it automatically learns important features such as edges, shapes, and textures directly from the raw pixel values. Traditional Machine Learning requires manual feature extraction, which is time-consuming and often less accurate. Therefore, Deep Learning models provide higher accuracy and better performance for image classification tasks.

Q3. A biological neuron fires only when incoming signals exceed a threshold. What is the artificial equivalent of this threshold mechanism in a neural network?

Answer:

In an Artificial Neural Network (ANN), the threshold mechanism is implemented using an activation function. The weighted sum of inputs and bias is passed through an activation function such as ReLU, Sigmoid, or Tanh to determine whether the neuron should produce an output.

Q4. You must choose between an ANN and a CNN for Fashion MNIST. What property of CNNs makes them better suited for image data than a plain ANN?

Answer:

CNNs are better suited for image data because they use convolutional layers to automatically detect spatial features such as edges, corners, textures, and shapes. They preserve the spatial relationship between pixels and require fewer parameters than a fully connected ANN, making them more accurate and efficient for image classification.

Q5. Deep Learning was theorized decades ago but became practical only recently. Name two factors that enabled the rise of modern Deep Learning.

Answer:

Two major factors that enabled the growth of modern Deep Learning are:

Powerful GPUs and TPUs, which significantly speed up neural network training.
Availability of large datasets (Big Data), allowing models to learn complex patterns more effectively.
